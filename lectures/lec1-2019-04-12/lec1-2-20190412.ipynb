- In real life we need to make decisions based on conditions. Conditions in real life are rarely binary and never deterministic.

### If Statement
```python
if <conditoin>:
    do sometning
```

````python
a, b = 5, 15
if b % a == 0:
    print("{} is divisible by {}".format(b, a))
````

### If ... else Statement
```python
if <complex conditoin>:
    do sometning
else:
    do something different
   
```

````python
a, b = 1, 15
if b % a == 0 and a != 1:
    print("{} is divisible by {}".format(b, a))
else:
    print("either {} is not divisable by {} or a==1 = True".format(b, a))
````

### if ... elif ... else
```python
if <condition>:
    do something
elif <another condition>:
    do something different
else:
    do something different no matter what
```


````python
#a, b = 5, 15
a, b = 1, 15

if a == 1:
    print("a==1 is an exception and will be skipped")
elif a == 0:
    print('a==0 will cause division by zero and is invalid')
elif b % a == 0:
    print("{} is divisible by {}".format(b, a))
else:
    print("{} is not divisible by {}".format(b, a))
````

## For loops
- We often need to loop over values and repeat the same operation over and over again. 
- For instance in a registration system, if pre-requisites are to be taken, we shall need to loop over students, who have requested registration and will check pre-requisite constraint satisfaction per student.

````python
candidates = ['Zeus', 'Hera', 'Melissa', 'Siegried', 
              'Venus', 'Dionysus', 'prometheus', 'Ares',
              'Apollo', 'Hermes']
gods = ['Zeus', 'Hera', 'Poseidon', 'Dermeter', 
        'Athena', 'Apollo', 'Artemis', 'Ares', 
        'Aphrodites', 'Hephaestus', 'Hermes']
accepted = []
rejected = []    
waiting_list = []
CAPACITY = 3
````

````python
#the class is only for olympian immortal gods. Check status and confirm or reject registration
accepted.clear()
rejected.clear()
for candidate in candidates:
    if candidate in gods:
        accepted.append(candidate)
    else:
        rejected.append(candidate)
print("accepted: {}".format(accepted))
print("rejected: {}".format(rejected))
````

````python
# There are only 3 seats avaialable, if there are more qualified candidates add them to a waiting list
accepted.clear()
rejected.clear()
waiting_list.clear()
for candidate in candidates:
    if candidate in gods:
        if len(accepted) < CAPACITY:
            accepted.append(candidate)
        else:
            waiting_list.append(candidate)
    else:
        rejected.append(candidate)
print("accepted: {}".format(accepted))
print("rejected: {}".format(rejected))
print("waiting list: {}".format(waiting_list))
````

### Breaking out of a for loop

````python
# There are only 3 seats avaialable, if there are more qualified candidates add them to a waiting list
CAPACITY = 4
accepted.clear()
rejected.clear()
c = candidates.copy() # we have used copy and not assignment
for i in range(len(candidates)):
    if len(accepted) >= CAPACITY:
        break
    if candidates[i] in gods:
        accepted.append(c.pop(0))
    else:
        rejected.append(c.pop(0))
rejected += c
(accepted, rejected)
````

## While Loop
- While loop is a conditional loop

````python
candidates = ['Zeus', 'Hera', 'Melissa', 'Siegried', 
              'Venus', 'Dionysus', 'prometheus', 'Ares',
              'Apollo', 'Hermes']
gods = ['Zeus', 'Hera', 'Poseidon', 'Dermeter', 
        'Athena', 'Apollo', 'Artemis', 'Ares', 
        'Aphrodites', 'Hephaestus', 'Hermes']
accepted = []
rejected = [] 
waiting_list = []
````

````python
CAPACITY = 3
accepted.clear()
rejected.clear()
waiting_list.clear()
c_counter = CAPACITY
while c_counter > 0 and len(candidates) > 0:
    if candidates[0] in gods:
        accepted.append(candidates.pop(0))
        c_counter -= 1
    else:
        rejected.append(candidates.pop(0))
waiting_list = candidates #we have used assignment and not copy
print(accepted, rejected, waiting_list)        
````

## Functions
- Whenever we can standardize and parametrize an operation, we would like to do that.
- Amongst benefits of using functions is modularization, ease of debugging, and simpler code base.
- Example 1:

- Example 2: We want to extend our Olympian university registration programme and provide courses for international student gods.

````python
def finalize_registeration(candidates, validation_list, capacity=5, external=True):
    accepted, rejected, waiting_list = [], [], []
    c_counter = capacity
    while c_counter > 0 and len(candidates) != 0:
        if candidates[0] in validation_list:
            accepted.append(candidates.pop(0))
            c_counter -= 1
        else:
            rejected.append(candidates.pop(0))
    waiting_list = candidates #we have used assignment and not copy
    if external:
        cap = capacity - len(accepted)
        for i in range(cap):
            if not len(rejected):
                break
            accepted.append(rejected.pop(0))
    return accepted, rejected, waiting_list      
````

````python
enrolled = ['Adam', 'Kathrin', 'Robert', 'Maria', 
              'Christopher', 'Golda', 'Matrin', 'Martha',
              'William', 'Suzanne']
passed_prereq = ['Adam', 'Martha', 'Kathrin', 'Robert', 
        'Daniel', 'Andrea', 'Manuel', 'Manuela', 
        'Gabriel', 'Gabriela', 'Paulo', 'Suzanne']

accepted, rejected, waiting_list = finalize_registeration(candidates=enrolled,
                                                           validation_list=passed_prereq,
                                                           capacity=8,
                                                           external=True)
(accepted, rejected, waiting_list)
````
