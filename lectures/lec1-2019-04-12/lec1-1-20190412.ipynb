## Parametric Algorithms
- Some of the algortihm have parameters which allows them to be more flexible and fit your data based on changes to parameters.
- Parametrics algorithms are better and generalization.
- Linear regression, for example, is a parametric algorithm.

### Linear Regression
- Remembering from high-school math, we know that in a two-dimensional space a straight line can be represented as a function, formalized as the following:

$$f:\mathcal{D} \rightarrow \mathcal{R}$$
$$y = f(x) = ax + b$$
- a and b are parameters of a two dimentional linear model.

```python
def plot():
    plt.scatter(data, x)
    s1 = plt.scatter(data, f(data, 1, 0), )
    s2 = plt.scatter(data, f(data, 1.25, 0))
    s3 = plt.scatter(data, f(data, 1.50, 0))
    s4 = plt.scatter(data, f(data, 1.25, 3))
    s5 = plt.scatter(data, f(data, -1.25, 8))
    plt.legend((s1, s2, s3, s4, s5),
              ('f(x)=x', 'f(x)=1.25x', 'f(x)=1.5x', 'f(x)=1.25x+3', 'f(x)= -1.25x+8'))
    plt.show()
```

```python
plot()
```

## Algorithmic Approach
- Challenge: in your groups find the best solution to the problem.

### Brute Force: 
 - List all solutions.
 - Discard all solutions with total weight larger than 15 kg.
 - Choose the solution that maximizes weight.

#### Issues:
- The solution space grows exponentially.

### Greedy Algorithm: 
 - Until the backpack is not full
  - Choose an item that is best fitting your goal.
 - *How did you determine what "best" was?*

#### Issues:
- There is no guarantee that we find the best answer as greedy algorithms focus on local optimization.

```python
def ackley2d(x, y, a=20, b=0.2, c=2*np.pi):
    d = 2
    sum1 = x**2 + y**2
    sum2 = np.cos(c*x) + np.cos(c*y)
    term1 = -a * np.exp(-b*np.sqrt(sum1/d))
    term2 = -np.exp(sum2/d)
    s = term1 + term2 + a + np.exp(1)
    return(s)

def cone_plot():
    fig = plt.figure(figsize=(10, 7))
    ax = fig.gca(projection='3d')
    x = np.linspace(-20, 20, 200)
    y = x
    X,Y = np.meshgrid(x, y)
    Z = ackley2d(X, Y)

    surf = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=plt.cm.coolwarm,
            linewidth=0, antialiased=False)

    #ax.set_zlim(0, 0.2)

    ax.zaxis.set_major_locator(plt.LinearLocator(10))
    ax.zaxis.set_major_formatter(plt.FormatStrFormatter('%.01f'))

    fig.colorbar(surf, shrink=0.5, aspect=7, cmap=plt.cm.coolwarm)
    plt.show()
    
def eggholder(x, y):
    return(-(y + 47)*np.sin(np.sqrt(np.abs(y+x/2+47))) - x*np.sin(np.sqrt(np.abs(x - (y+47)))))

def egg_manifold():
    #fig = plt.figure(figsize=(10, 7), facecolor='#1F1F1F') #Black
    fig = plt.figure(figsize=(10, 7))

    #ax = fig.gca(projection='3d', facecolor='#1F1F1F')
    ax = fig.gca(projection='3d')

    x = np.linspace(-256, 256, 200)
    y = x
    X,Y = np.meshgrid(x, y)
    Z = eggholder(X, Y)

    #surf = ax.plot_surface(X, Y, Z, rstride=1, cstride=1,cmap=plt.cm.coolwarm,
    #        linewidth=0, antialiased=False, facecolor='#1F1F1F')
    surf = ax.plot_surface(X, Y, Z, rstride=1, cstride=1,cmap=plt.cm.coolwarm,
            linewidth=0, antialiased=False)

    #ax.set_zlim(0, 0.2)

    ax.zaxis.set_major_locator(plt.LinearLocator(10))
    ax.zaxis.set_major_formatter(plt.FormatStrFormatter('%.02f'))

    plt.tick_params(axis='both', which='major', labelsize=8)
    ax.tick_params(labelsize=8, color='red') 
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_zticks([])
    ax.grid(False)
    ax.w_xaxis.set_pane_color((0.0, 0.0, 0.0, 0.0))
    ax.w_yaxis.set_pane_color((0.0, 0.0, 0.0, 0.0))
    ax.w_zaxis.set_pane_color((0.0, 0.0, 0.0, 0.0))
    ax.grid(True)

    #fig.colorbar(surf, shrink=0.5, aspect=7, cmap=plt.cm.coolwarm)

    plt.show()
```

## Optimization and Convexity
$$f:\mathbb{R}^n \rightarrow \mathbb{R}\text{ is convex if and only if } \forall x_1,x_2 \epsilon \mathbb{R}^n\ and\  \lambda \epsilon[0, 1]\\ f(\lambda x_1 + (1-\lambda)x_2) \leq \lambda f(x_1)+(1-\lambda)f(x_2)$$

Non-Convex optimization problem may have multiple feasible regions and multiple locally optimal points within each region.
